```python
from selenium.webdriver import Chrome
import urllib.request as r
import time
```

```python
def openUrl():
    '''打开网站'''
    web= Chrome(executable_path='chromedriver.exe')
    url='https://ics.autohome.com.cn/passport/'
    web.get(url)
    time.sleep(1)
    return web
```
